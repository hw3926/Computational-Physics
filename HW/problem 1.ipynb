```python
import numpy as np
import matplotlib.pyplot as plt


def forward_diff(f, x, h):
    return (f(np.float32(x + h)) - f(np.float32(x))) / h

def central_diff(f, x, h):
    return (f(np.float32(x + h)) - f(np.float32(x - h))) / (2*h)

def extrapolated_diff(f, x, h):
    D_h = central_diff(f, x, 2*h)
    D_h2 = central_diff(f, x, h)
    return (4*D_h2 - D_h) / 3


hh_log = np.logspace(-8, 0, 300, dtype=np.float32)

x=0.1
deriv_cos_forward = forward_diff(np.cos, x, hh_log)
deriv_cos_central = central_diff(np.cos, x, hh_log)
deriv_cos_extrapolated = extrapolated_diff(np.cos, x, hh_log)
err_cos_forward = np.abs((deriv_cos_forward + np.sin(x))/np.sin(x))
err_cos_central = np.abs((deriv_cos_central + np.sin(x))/np.sin(x))
err_cos_extrapolated = np.abs((deriv_cos_extrapolated + np.sin(x))/np.sin(x))
plt.loglog(hh_log, err_cos_forward,label = 'cos forward')
plt.loglog(hh_log, err_cos_central,label = 'cos central')
plt.loglog(hh_log, err_cos_extrapolated,label = 'cos extrapolated')
plt.legend()
plt.xlabel("Step size h")
plt.ylabel("Relative error")
plt.title("Derivative of cos$(0.1)$")
plt.grid(True, which="major", ls="--", alpha=0.6)
plt.show()

x=10
deriv_cos_forward = forward_diff(np.cos, x, hh_log)
deriv_cos_central = central_diff(np.cos, x, hh_log)
deriv_cos_extrapolated = extrapolated_diff(np.cos, x, hh_log)
err_cos_forward = np.abs((deriv_cos_forward + np.sin(x))/np.sin(x))
err_cos_central = np.abs((deriv_cos_central + np.sin(x))/np.sin(x))
err_cos_extrapolated = np.abs((deriv_cos_extrapolated + np.sin(x))/np.sin(x))
plt.loglog(hh_log, err_cos_forward,label = 'cos forward')
plt.loglog(hh_log, err_cos_central,label = 'cos central')
plt.loglog(hh_log, err_cos_extrapolated,label = 'cos extrapolated')
plt.legend()
plt.xlabel("Step size h")
plt.ylabel("Relative error")
plt.title("Derivative of cos$(10)$")
plt.grid(True, which="major", ls="--", alpha=0.6)
plt.show()


x = 0.1
deriv_exp_forward = forward_diff(np.exp, x, hh_log)
deriv_exp_central = central_diff(np.exp, x, hh_log)
deriv_exp_extrapolated = extrapolated_diff(np.exp, x, hh_log)

exact_exp = np.exp(x)

err_exp_forward = np.abs((deriv_exp_forward - exact_exp) / exact_exp)
err_exp_central = np.abs((deriv_exp_central - exact_exp) / exact_exp)
err_exp_extrapolated = np.abs((deriv_exp_extrapolated - exact_exp) / exact_exp)

plt.loglog(hh_log, err_exp_forward, label='exp forward')
plt.loglog(hh_log, err_exp_central, label='exp central')
plt.loglog(hh_log, err_exp_extrapolated, label='exp extrapolated')

plt.legend()
plt.xlabel("Step size h")
plt.ylabel("Relative error")
plt.title("Derivative of exp$(0.1)$")
plt.grid(True, which="major", ls="--", alpha=0.6)
plt.show()


x = 10
deriv_exp_forward = forward_diff(np.exp, x, hh_log)
deriv_exp_central = central_diff(np.exp, x, hh_log)
deriv_exp_extrapolated = extrapolated_diff(np.exp, x, hh_log)

exact_exp = np.exp(x)

err_exp_forward = np.abs((deriv_exp_forward - exact_exp) / exact_exp)
err_exp_central = np.abs((deriv_exp_central - exact_exp) / exact_exp)
err_exp_extrapolated = np.abs((deriv_exp_extrapolated - exact_exp) / exact_exp)

plt.loglog(hh_log, err_exp_forward, label='exp forward')
plt.loglog(hh_log, err_exp_central, label='exp central')
plt.loglog(hh_log, err_exp_extrapolated, label='exp extrapolated')

plt.legend()
plt.xlabel("Step size h")
plt.ylabel("Relative error")
plt.title("Derivative of exp$(10)$")
plt.grid(True, which="major", ls="--", alpha=0.6)
plt.show()

        
```
