```python
# dow_a.py
import numpy as np
import matplotlib.pyplot as plt

# Load data: one column of daily closing values
data = np.loadtxt("dow.txt")

plt.figure(figsize=(10, 4))
plt.plot(data, color="navy", linewidth=0.8)
plt.xlabel("Day index")
plt.ylabel("Dow Jones closing value")
plt.title("Dow Jones Industrial Average (20062007)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# dow_b.py
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("dow.txt")
N = len(data)

# Real FFT
coeffs = np.fft.rfft(data)

# Plot magnitude spectrum
freqs = np.fft.rfftfreq(N, d=1)  # frequency bins (in cycles/day)
plt.figure(figsize=(10, 4))
plt.plot(freqs, np.abs(coeffs), color="darkred")
plt.xlabel("Frequency (cycles/day)")
plt.ylabel("Magnitude of Fourier coefficients")
plt.title("Fourier Spectrum of Dow Data")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# dow_c.py
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("dow.txt")
N = len(data)

coeffs = np.fft.rfft(data)

# Keep only first 10% of coefficients
cutoff = int(0.1 * len(coeffs))
filtered = np.zeros_like(coeffs)
filtered[:cutoff] = coeffs[:cutoff]

# Inverse transform
smoothed = np.fft.irfft(filtered, n=N)

plt.figure(figsize=(10, 4))
plt.plot(data, label="Original", color="navy", linewidth=0.8)
plt.plot(smoothed, label="Smoothed (10%)", color="darkorange", linewidth=1.2)
plt.xlabel("Day index")
plt.ylabel("Dow Jones closing value")
plt.title("Fourier Filtering (10% coefficients)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# dow_e.py
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("dow.txt")
N = len(data)

coeffs = np.fft.rfft(data)

# Keep only first 2% of coefficients
cutoff = int(0.02 * len(coeffs))
filtered = np.zeros_like(coeffs)
filtered[:cutoff] = coeffs[:cutoff]

smoothed = np.fft.irfft(filtered, n=N)

plt.figure(figsize=(10, 4))
plt.plot(data, label="Original", color="navy", linewidth=0.8)
plt.plot(smoothed, label="Smoothed (2%)", color="crimson", linewidth=1.2)
plt.xlabel("Day index")
plt.ylabel("Dow Jones closing value")
plt.title("Fourier Filtering (2% coefficients)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
